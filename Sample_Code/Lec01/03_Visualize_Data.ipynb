```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ tập tin csv
df = pd.read_csv('03_Customer Purchasing Behaviors.csv')


# 1. Biểu đồ phân bố độ tuổi
plt.figure(figsize=(8, 5))
plt.hist(df['age'], bins=10, color='blue', edgecolor='black')
plt.title('Phân bố độ tuổi của khách hàng')
plt.xlabel('Độ tuổi')
plt.ylabel('Số lượng khách hàng')
plt.show()

# 2. Biểu đồ thanh về tổng thu nhập hàng năm theo khu vực
income_by_region = df.groupby('region')['annual_income'].sum()
plt.figure(figsize=(8, 5))
income_by_region.plot(kind='bar', color='green', edgecolor='black')
plt.title('Tổng thu nhập hàng năm theo khu vực')
plt.xlabel('Khu vực')
plt.ylabel('Tổng thu nhập hàng năm')
plt.show()

```
